# Lời nói đầu: Giới thiệu về Phân tích Chuỗi Thời Gian (Time Series Analysis)

Cuốn sổ tay (handbook) này trình bày toàn diện về phân tích và dự báo chuỗi thời gian, đi từ những phương pháp cơ bản nhất cho đến các kỹ thuật hiện đại (state-of-the-art). Sổ tay được xây dựng bằng Python và được thiết kế sao cho người đọc vừa có thể học lý thuyết, vừa có thể áp dụng trực tiếp vào các bài toán thực tế. Mặc dù các chương được xây dựng theo hướng độc lập, người đọc vẫn được khuyến nghị nên đọc những chương đầu tiên để có thể hiểu và đánh giá đầy đủ hơn các chương về sau. Ngoài ra,
__[bộ dữ liệu khí hậu Jena](https://www.kaggle.com/datasets/trnvquanghuy/jena-climate)__ được sử dụng xuyên suốt trong nhiều chương, kèm theo bản tổng kết hiệu năng của các mô hình ở phần cuối.

Cấu trúc của sổ tay như sau:
ở phần đầu, các phương pháp dự báo cổ điển được trình bày chi tiết. Phần giữa tập trung vào các phương pháp dự báo động lực, cũng như nhân quả và tương quan — những chủ đề đặc biệt quan trọng để hiểu rõ sự phức tạp của bài toán dự báo chuỗi thời gian. Cuối cùng, phần cuối cung cấp một cái nhìn khái quát về các xu hướng hiện tại và những bài toán còn mở trong lĩnh vực dự báo và mô hình hóa chuỗi thời gian.

Mục tiêu của sổ tay này là đóng vai trò như một cẩm nang thực hành (practitioner’s guide) cho dự báo, giúp người đọc hiểu rõ hơn các mối quan hệ giữa các tín hiệu. Tài liệu được viết cho đối tượng độc giả có nền tảng vững về Thống kê và Toán học, cũng như kiến thức Python cơ bản. Việc quen thuộc với các phương pháp Học máy (Machine Learning) là một lợi thế, đặc biệt đối với các chương về sau.
